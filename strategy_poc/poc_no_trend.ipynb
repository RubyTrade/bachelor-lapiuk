# POC: baseline (no regime routing)

Same **futures long/short** and **combined imbalance** strategies, but **without** switching tactics by trend (`regime_meta` is not run here — see `poc_trend.ipynb`).

Use to compare “single tactic all year” vs “regime” in the other notebook.

```python
import os

from poc_backtest_core import (
    TP_SL_PRESETS,
    IMBALANCE_THRESHOLDS,
    MEAN_DELTAS,
    MOMENTUM_THRESHOLDS,
    BREAKOUT_LOOKBACKS,
    analytics_full,
    clear_reports,
    mean_strategy_futures,
    momentum_strategy_futures,
    breakout_strategy_futures,
    momentum_with_imbalance_confirm,
    mean_with_imbalance_confirm,
    run_backtest_yearly,
)

data_folder = "data"
symbol = "BTCUSDT"
year = 2025
months_full = list(range(1, 13))

_sample = os.path.join(data_folder, f"{symbol}-aggTrades-{year}-01.parquet")
if not os.path.isfile(_sample):
    print(f"No data at {_sample!r}")
```

```python
clear_reports()

if os.path.isfile(_sample):
    for _preset, (_tp, _sl) in TP_SL_PRESETS.items():
        for _thr in IMBALANCE_THRESHOLDS:
            run_backtest_yearly(
                "mom+imb",
                momentum_with_imbalance_confirm,
                data_folder=data_folder,
                symbol=symbol,
                year=year,
                months=months_full,
                eval_label="no_trend_full",
                threshold=_thr,
                momentum_threshold=0.001,
                tp=_tp,
                sl=_sl,
            )
            run_backtest_yearly(
                "mean+imb",
                mean_with_imbalance_confirm,
                data_folder=data_folder,
                symbol=symbol,
                year=year,
                months=months_full,
                eval_label="no_trend_full",
                threshold=_thr,
                delta=0.002,
                tp=_tp,
                sl=_sl,
            )
        for _delta in MEAN_DELTAS:
            run_backtest_yearly(
                "mean",
                mean_strategy_futures,
                data_folder=data_folder,
                symbol=symbol,
                year=year,
                months=months_full,
                eval_label="no_trend_full",
                delta=_delta,
                tp=_tp,
                sl=_sl,
            )
        for _mt in MOMENTUM_THRESHOLDS:
            run_backtest_yearly(
                "momentum",
                momentum_strategy_futures,
                data_folder=data_folder,
                symbol=symbol,
                year=year,
                months=months_full,
                eval_label="no_trend_full",
                momentum_threshold=_mt,
                tp=_tp,
                sl=_sl,
            )
        for _lb in BREAKOUT_LOOKBACKS:
            run_backtest_yearly(
                "breakout",
                breakout_strategy_futures,
                data_folder=data_folder,
                symbol=symbol,
                year=year,
                months=months_full,
                eval_label="no_trend_full",
                lookback=_lb,
                tp=_tp,
                sl=_sl,
            )

    analytics_full("poc_backtest_leaderboard_no_trend.csv")
```
